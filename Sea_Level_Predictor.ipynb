```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

### Read data from file

```python
df = pd.read_csv("epa-sea-level.csv")
df
```

### Create scatter plot

```python
fig, ax = plt.subplots()
ax = plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
```

### Create first line of best fit

```python
lin = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
ax=plt.plot(df['Year'], lin.slope*df['Year'] + lin.intercept)
```

```python
ax = plt.plot(np.arange(2000,2050), lin.slope*np.arange(2000,2050) + lin.intercept)
```

### Create second line of best fit

```python
ax = plt.plot(np.arange(2000, 2050), lin.slope *
              np.arange(2000, 2050) + lin.intercept)
```

### Add labels and title

```python
plt.xlabel("Year")
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
```

### Combining everything into a single function

```python
def draw_plot():
    df = pd.read_csv("epa-sea-level.csv")

    fig, ax = plt.subplots()
    ax = plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])

    lin = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    ax = plt.plot(df['Year'], lin.slope*df['Year'] + lin.intercept)
    
    ax = plt.plot(np.arange(2000, 2050), lin.slope *
                  np.arange(2000, 2050) + lin.intercept)

    plt.xlabel("Year")
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
    return plt.gca()
```

### The Plot

```python
draw_plot()
```
